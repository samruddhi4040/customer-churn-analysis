```python
df['Churn']=df['Churn'].map({'Yes':1,'No':0})
```

```python
df['Churn'].value_counts()
```

```python
df['Churn']=pd.to_numeric(df['Churn'], errors='coerce')
```

```python
df['Churn'].isnull().sum()
```

```python
#Feature Engineering
```

```python
#Creating tenure groups 
```

```python
df['TenureGroup']=pd.cut(df['tenure'],
                         bins=[0,12,24,48,72],
                         labels=['0-1 yr', '1-2 yr','2-4 yr', '4-6 yr'])
```

```python
df[['tenure','TenureGroup']].head()
```

```python
df['ChargeLevel']=pd.cut(df['MonthlyCharges'],
                         bins=3,
                         labels=['Low','Medium','High'])
```

```python
df[['MonthlyCharges','ChargeLevel']].head()
```

```python
df.to_csv("cleaned_churn.csv",index=False)
```
